## Monkeypox_cases analysis

Monkeypox is a rare disease caused by the monkeypox virus.The first human case of monkeypox was recorded in 1970. Prior to the 2022 outbreak, monkeypox had been reported in people in several central and western African countries.
People with monkeypox get a rash that may be located on or near the genitals and could be on other areas like the hands, feet, chest, face, or mouth.

The rash will go through several stages, including scabs, before healing.
The rash can initially look like pimples or blisters and may be painful or itchy.Other symptoms of monkeypox can include: Fever,Chills,Swollen lymph nodes,Exhaustion,Muscle aches and backache,Headache,Respiratory symptoms.

### Total number of cases in each country

### Top 5 countries with highest no of monkeypox cases

```python
#converting date column into datetime format
df["Date"]=pd.to_datetime(df["Date"],format='%y-%m-%d',errors='ignore')
#splitting date column into year , month and day column 
df[['year','month','day']]=df['Date'].str.split('-',expand=True)
#counting no of cases for each country 
count=df.groupby(df.year)[df.columns].sum().max().sort_values(ascending=False)
#top five countries and their total no of cases
country=count.index[:5]
value=count.values[:5]
print(country,value)
def add_value_label(x_list,y_list):
    for i in range(1,len(x_list)+1):
        plt.annotate(y_list[i-1],(i-1,y_list[i-1]/2),ha="center")

#plotting the bar graph
fig=plt.bar(country,value)
add_value_label(country,value)
plt.title('Top 5 countries with highest cases of monkeypox in 2022')
plt.ylabel(" Total cases ",fontsize=18)
plt.xlabel("Country",fontsize=18)
plt.show()
type(count)
```

###  Total cases worldwide per day

```python
df1 = df.iloc[:, 1:78].sum(axis=1)
df1.index = df.iloc[:, 0]
plt.figure(figsize=(20, 20))
plt.title=("Total cases world wide per day")
plt.ylabel=("Cases")
plt.xlabel=("Date")
plt.plot(df1.index,df1)
plt.xticks(rotation=90)

plt.show()
```

### Monthly count of  cases from jan 2022 to july 2022

```python
country_names=list(df.columns)[1:78]
l=len(country_names)
for i in country_names:
    monthly_count=df.groupby(df.month)[str(i)].sum()
    m=monthly_count.index
    a=['Jan','Feb','Mar','Apr','May','Jun','July']
    v=monthly_count.values
    fig = plt.bar(a,v)
    add_value_label(a,v)
    plt.xlabel("Month",size=12)
    plt.ylabel("Total cases",size=12)
    plt.suptitle(i,size=20)
    plt.show()
```
