### Top 5 countries with highest no of monkeypox cases

```python
df["Date"]=pd.to_datetime(df["Date"],format='%y-%m-%d',errors='ignore')
df[['year','month','day']]=df['Date'].str.split('-',expand=True)
count=df.groupby(df.year)[df.columns].sum().max().sort_values(ascending=False)
country=count.index[:5]
value=count.values[:5]
print(country,value)
def add_value_label(x_list,y_list):
    for i in range(1,len(x_list)+1):
        plt.annotate(y_list[i-1],(i-1,y_list[i-1]/2),ha="center")

fig=plt.bar(country,value)
add_value_label(country,value)
plt.title('Top 5 countries with highest cases of monkeypox in 2022')
plt.ylabel(" Total cases ",fontsize=18)
plt.xlabel("Country",fontsize=18)
plt.show()
type(count)
```

```python
plt.plot(country,value)
plt.title('Top 5 countries with highest cases of monkeypox in 2022')
plt.ylabel(" Total cases ",size=18)
plt.xlabel("Country",size=18)
plt.show()
```

```python
country_names=list(df.columns)[1:78]
l=len(country_names)
for i in country_names:
    monthly_count=df.groupby(df.month)[str(i)].sum()
    m=monthly_count.index
    a=['Jan','Feb','Mar','Apr','May','Jun','July']
    v=monthly_count.values
    fig = plt.bar(a,v)
    add_value_label(a,v)
    plt.xlabel("Month",size=12)
    plt.ylabel("Total cases",size=12)
    plt.suptitle(i,size=20)
    plt.show()
```

```python
for country in country_names:
    plt.figure(figsize=(15, 5))
    plt.plot(df.month , df[country], color = 'red' ,marker='o',markerfacecolor="r",markeredgecolor="b")
 
    plt.xticks(np.arange(2021,2022) ,rotation=90)
    plt.suptitle(country, fontsize=18)
    plt.ylabel("MonkeyPox",size=12)
    plt.xlabel("monthly trend in 2022",size=12)
    plt.title("frequency in month")
    plt.show()
```
